```python
import nltk
import random
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('punkt')
```

```python
customer_data = {
    "greetings": {
        "patterns": ["hi", "hello", "hey", "good morning", "good evening"],
        "responses": ["Hello! How can I help you?", "Hi there! How may I assist you?"]
    },
    "order_status": {
        "patterns": ["where is my order", "order status", "track my order"],
        "responses": ["Please provide your order ID to check the status."]
    },
    "refund": {
        "patterns": ["refund", "money back", "return product"],
        "responses": ["Refunds are processed within 5–7 business days."]
    },
    "working_hours": {
        "patterns": ["working hours", "open time", "office hours"],
        "responses": ["We are open from 9 AM to 6 PM, Monday to Friday."]
    },
    "contact": {
        "patterns": ["contact support", "customer care number", "helpdesk"],
        "responses": ["You can contact us at support@example.com."]
    }
}
```

```python
patterns = []
responses = []

for intent in customer_data:
    for pattern in customer_data[intent]["patterns"]:
        patterns.append(pattern)
        responses.append(random.choice(customer_data[intent]["responses"]))
```

```python
def preprocess(text):
    text = text.lower()
    text = text.translate(str.maketrans("", "", string.punctuation))
    return text
```

```python
vectorizer = TfidfVectorizer(preprocessor=preprocess)
X = vectorizer.fit_transform(patterns)
```

```python
def chatbot_response(user_input):
    user_input = preprocess(user_input)
    user_vec = vectorizer.transform([user_input])
    similarity = cosine_similarity(user_vec, X)
    idx = similarity.argmax()

    if similarity[0][idx] > 0.3:
        return responses[idx]
    else:
        return "Sorry, I didn't understand that. Please try again."
```

```python
print("Customer Support Chatbot (type 'exit' to stop)")

while True:
    user = input("You: ")
    if user.lower() == "exit":
        print("Bot: Thank you for contacting us!")
        break
    print("Bot:", chatbot_response(user))
```
