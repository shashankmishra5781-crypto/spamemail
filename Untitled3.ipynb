```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
data = pd.read_csv('spam.csv', encoding = 'latin-1')
```

```python
#drop duplicates
data.drop_duplicates(inplace=True)
```

```python
print(data.head())
```

```python
mess = data['Message']
lab = data['Label']
```

```python
mess_train, mess_test, lab_train, lab_test = train_test_split(mess, lab, test_size=0.2)
```

```python
cv = CountVectorizer(stop_words = 'english')
```

```python
cv.fit_transform(mess_train)
features = cv.fit_transform(mess_train)
```

```python
#creating our model
model = MultinomialNB()
```

```python
model.fit(features, lab_train)
```

```python
#Test our model
features_test = cv.transform(mess_test)
```

```python
print(model.score(features_test, lab_test))
```

```python
#predicting in real time
message = cv.transform(['congratulation,you have won a lottery']).toarray()
result = model.predict(message)
print(result)
```
